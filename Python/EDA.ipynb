```python
# import needed libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Data from CSV
file_path = r"cleaned_agriculture_data.csv"
df = pd.read_csv(file_path)
```

```python
df
```

---
## EDA Visuals
#### 1. Top 7 Rice Producing States (Bar Plot)



```python
top_rice_states = df.groupby('state_name')['rice_production'].sum().nlargest(7)
top_rice_states.plot(kind='bar', title='Top 7 Rice Producing States')
```

---
#### 2. Top 5 Wheat Producing States (Bar Chart + Pie Chart for % Contribution)


```python
top_wheat_states = df.groupby('state_name')['wheat_production'].sum().nlargest(5)
# Bar chart
top_wheat_states.plot(kind='bar', title='Top 5 Wheat Producing States')
```

```python
# Pie chart for % contribution
top_wheat_states.plot(kind='pie', autopct='%1.1f%%', title='Wheat Production % by Top 5 States')
```

---
#### 3. Oilseed Production by Top 5 States

```python
top_oilseed_states = df.groupby('state_name')['oilseeds_production'].sum().nlargest(5)
top_oilseed_states.plot(kind='bar', title='Top 5 Oilseed Producing States')
```

---
#### 4. Top 7 Sunflower Producing States

```python
top_sunflower_states = df.groupby('state_name')['sunflower_production'].sum().nlargest(7)
top_sunflower_states.plot(kind='bar', title='Top 7 Sunflower Producing States')
```

---
#### 5. Indiaâ€™s Sugarcane Production Over 50 Years (Line Plot)

```python
sugarcane_over_time = df.groupby('year')['sugarcane_production'].sum()
sugarcane_over_time.plot(kind='line', title='India Sugarcane Production (Last 50 Years)')
```

---
#### 6. Rice vs. Wheat Production (Last 50 Years)

```python
production_compare = df.groupby('year')[['rice_production', 'wheat_production']].sum()
production_compare.plot(kind='line', title='Rice vs Wheat Production (Last 50 Years)')
```

---
#### 7. Rice Production by West Bengal Districts



```python
wb_rice = df[df['state_name'].str.lower() == 'west bengal'].groupby('dist_name')['rice_production'].sum().sort_values(ascending=False)
wb_rice.plot(kind='bar', title='Rice Production by West Bengal Districts')
```

---
#### 8. Top 10 Wheat Production Years from UP



```python
up_wheat = df[df['state_name'].str.lower() == 'uttar pradesh'].groupby('year')['wheat_production'].sum().nlargest(10)
up_wheat.plot(kind='bar', title='Top 10 Wheat Production Years - Uttar Pradesh')
```

---
#### 9. Millet Production (Last 50 Years)

```python
df['total_millet'] = df[['pearl_millet_production', 'finger_millet_production']].sum(axis=1)
millet_trend = df.groupby('year')['total_millet'].sum()
millet_trend.plot(kind='line', title='Millet Production Over 50 Years')
```

---
#### 10. Sorghum Production (Kharif and Rabi) by Region

```python
sorghum_by_state = df.groupby('state_name')[['kharif_sorghum_production', 'rabi_sorghum_production']].sum()
sorghum_by_state.plot(kind='bar', stacked=True, title='Sorghum Production by State (Kharif & Rabi)')
```

----
#### 11. Top 7 States for Groundnut Production


```python
top_groundnut_states = df.groupby('state_name')['groundnut_production'].sum().nlargest(7)
top_groundnut_states.plot(kind='bar', title='Top 7 Groundnut Producing States')
```

----
#### 12. Soybean Production by Top 5 States and Yield Efficiency

```python
top_soybean = df.groupby('state_name')[['soyabean_production', 'soyabean_yield']].sum().nlargest(5, 'soyabean_production')
top_soybean[['soyabean_production']].plot(kind='bar', title='Top 5 Soybean Producing States')
top_soybean['yield_efficiency'] = top_soybean['soyabean_yield'] / 1000
top_soybean['yield_efficiency'].plot(kind='line', secondary_y=True, color='red', marker='o')
```

---
#### 13. Oilseed Production in Major States

```python
major_oilseed_states = df.groupby('state_name')['oilseeds_production'].sum().nlargest(10)
major_oilseed_states.plot(kind='bar', title='Oilseed Production in Major States')
```

---
#### 14. Impact of Area Cultivated on Production (Rice, Wheat, Maize)

```python
# Set style
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=False)

# Plot for Rice
sns.regplot(ax=axes[0], data=df, x='rice_area', y='rice_production', color='green')
axes[0].set_title('Rice: Area vs Production')
axes[0].set_xlabel('Area Cultivated (1000 ha)')
axes[0].set_ylabel('Production (1000 tons)')

# Plot for Wheat
sns.regplot(ax=axes[1], data=df, x='wheat_area', y='wheat_production', color='orange')
axes[1].set_title('Wheat: Area vs Production')
axes[1].set_xlabel('Area Cultivated (1000 ha)')
axes[1].set_ylabel('Production (1000 tons)')

# Plot for Maize
sns.regplot(ax=axes[2], data=df, x='maize_area', y='maize_production', color='blue')
axes[2].set_title('Maize: Area vs Production')
axes[2].set_xlabel('Area Cultivated (1000 ha)')
axes[2].set_ylabel('Production (1000 tons)')

# Adjust layout
plt.suptitle('Impact of Area Cultivated on Production (Rice, Wheat, Maize)', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
